#9. Create a data frame that computes the average of ages in each countries?

Hint: Use groupby, select the age column and aggregate using mean.

```python
grouped_by_country = data.groupby('nationality')['age']
avg_ages_by_country = grouped_by_country.mean()
average_ages = avg_ages_by_country.reset_index()
average_ages
```
